```python
left=[]
right=[]
```

```python
with open("input.txt") as f:
    for line in f:
        a,b = map(int, line.split())
        left.append(a)
        right.append(b)
```

```python
right
```

```python
left.sort()
right.sort()
```

```python
right
```

```python
sum(abs(a-b) for a,b in zip(left,right))
```

```python
countleft= {a:right.count(a) for a in left}
```

```python
countleft.items()
```

```python
sum(a*n for a,n in countleft.items())
```
